### Unstabilized specimens at 70,140 and 190 kPa

```python
#04, 05 and 09 Unstabilized 
fig=plt.figure(figsize=(15,4))
ax1 = plt.subplot(121)
plt.plot(unreinf_04['cycle'],unreinf_04['surf_disp'],label='70 kPa')
plt.plot(unreinf_05['cycle'],unreinf_05['surf_disp'],label='140 kPa')
plt.plot(unreinf_09['cycle'],unreinf_09['surf_disp'],label='190 kPa')
plt.xlabel('No of cycles')
plt.ylabel('Vertical deformation (m)')
plt.title('Effect of loading stress - Unstabilized')
plt.legend()
plt.axis([0,500,-0.12,0])

ax2 = plt.subplot(122)
plt.plot(unreinf_04['cycle'],unreinf_04['porosity'],label='70 kPa')
plt.plot(unreinf_05['cycle'],unreinf_05['porosity'],label='140 kPa')
plt.plot(unreinf_09['cycle'],unreinf_09['porosity'],label='190 kPa')
plt.xlabel('No of cycles')
plt.ylabel('Porosity')
plt.title('Effect of loading stress - Unstabilized')
plt.axis([0,500,0.4,0.42])
plt.legend()
plt.show()
fig.savefig('effect_of_stress_unstab.png')

print('initial porosities:',unreinf_04['porosity'][0],unreinf_05['porosity'][0],unreinf_09['porosity'][0])
```

```python
#04, 05 and 09 Unstabilized 
fig=plt.figure(figsize=(15,12))
ax1 = plt.subplot(211)
plt.plot(unreinf_09['cycle'],unreinf_09['surf_disp'],c='b',label='190 kPa-Unstab')
plt.plot(reinf_02['cycle'],reinf_02['surf_disp'],'--',c='b',label='190 kPa-25mmGG')
plt.xlabel('No of cycles')
plt.ylabel('Vertical deformation (m)')
plt.title('Effect of loading stress and Stabilization')
plt.legend(loc=1)
plt.axis([0,600,-0.12,0])

ax2 = plt.subplot(212)
plt.plot(unreinf_09['cycle'],unreinf_09['porosity'],c='b',label='190 kPa-Unstab')
plt.plot(reinf_02['cycle'],reinf_02['porosity'],'--',c='b',label='190 kPa-25mmGG')
plt.xlabel('No of cycles')
plt.ylabel('Porosity')
plt.axis([0,600,0.4,0.43])
plt.legend()

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=.4)
plt.show()
fig.savefig('gg_effect_190kpa.png')
```
